#### Problem Statement 2:

The college bookstore tells prospective students that the average cost of its
textbooks is Rs. 52 with a standard deviation of Rs. 4.50. A group of smart statistics students thinks that the average cost is higher. To test the bookstore’s claim against their alternative, the students will select a random sample of size 100. Assume that the mean from their random sample is Rs. 52.80. Perform a hypothesis test at the 5% level of significance and state your decision.

```python
alpha=0.05
pop_mean=52
pop_std=4.5
n=100
samp_mean=52.80
```

```python
#The population variance is given and the sample size>30 we can use the z-test
std_err = pop_std/(np.sqrt(n))
z_stat = (samp_mean-pop_mean)/std_err
p=2*(1-(st.norm.cdf(np.absolute(z_stat))))
```

```python
if p<alpha:
    print("Null Hypothesis is rejected, 𝜇!=52")
else:
    print("Fail to reject null Hypothesis,  𝜇=52")
```

Therefore, we fail to reject null hypothesis and the average cost is less than or equal to Rs. 52

#### Problem Statement 15:
Comparisons of two population proportions when the hypothesized difference is zero
Carry out a two-tailed test of the equality of banks’ share of the car loan market in
1980 and 1995.

Population 1: 1980

n1 = 1000

x1 = 53

p 1 = 0.53

Population 2: 1985

n2 = 100

x2 =43

p2 = 0.53

```python
def z_test_pop_proportions(p1,p2,se):
    return ((p1-p2)/se)
def sample_error_pop_proportions(p1,p2,n1,n2):
    return np.sqrt(((p1*(1-p1))/n1)+((p2*(1-p2))/n2))
alpha=0.05
p1=53/1000
p2=43/100
```

```python
z=z_test_pop_proportions(p1=p1,p2=p2,se=sample_error_pop_proportions(p1=p1,p2=p2,n1=1000,n2=100))
p=2*(1-st.norm.cdf(x=np.absolute(z)))
```

```python
if p<alpha:
    print("Null Hypothesis is rejected(Banks share of the car loan market in 1980 and 1995 is not equal(P1!=P2))")
else:
    print("Fail to reject null hypothesis(Banks share of the car loan market in 1980 and 1995 is equal(P1=P2))")
```
