```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
df = datasets.load_iris()
X = df.data[:, :-1]
y = df.target
```

```python
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
red_x = decomposition.PCA(n_components=3).fit_transform(df.data)
ax.scatter(red_x[:, 0], red_x[:,1], red_x[:,2], c=y, 
           cmap=plt.cm.Set1, edgecolor='k', s=100)
ax.set_title("Transformed 3D Iris data")
ax.set_xlabel("1st Category")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd Category")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd Category")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
